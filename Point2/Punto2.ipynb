```python
%pylab inline
import numpy as np
from scipy.fftpack import fft, fftfreq
import scikits.audiolab as audio
import matplotlib.pyplot as plt
```

A continuación se lee la señal de la grabación y se grafica.

```python
signal, sample_frequency, encoding = audio.wavread("original_hw5.wav")
plot((signal[:,0]+signal[:,1])/2)
sig=((signal[:,0]+signal[:,1])/2)
figure()
plot(signal)
```

Ahora se realiza la transformada rápida de Fourier de la señar original y se hace la normalización

```python
points = len(sig)
fft_signal = fft(sig) / points 
freq = fftfreq(points, 1) 
plt.plot(freq,np.abs(fft_signal))
```

A continuación se graficará el espectro de potencias, el cual se define como la norma al cuadrado de cada una de las componentes de Fourier a una frecuencia determinada.

```python
half_n = np.ceil(points)
fft_signal_half = (2.0)*fft_signal[:half_n]
freq_half = freq[:half_n]
print size(freq_half), size(log10(np.absolute(fft_signal_half)))
scatter(freq_half*2, log10(np.absolute(fft_signal_half)**2)) 
#plt.xlim([0.1,0.5])
#plt.ylim([-6,-1])
```

### White noise

El ruido blanco se caracteriza por tener la misma distribución de potencia (promedio de energía de una señal en un periodo de tiempo) para todas las frecuencias.

La siguiente gráfica representa el espectro plano del ruido blanco

```python
whitespect = fft_signal_half/np.absolute(fft_signal_half)
plot(absolute(whitespect))
```

Ahora se halla la transformada inversa de está señal y se plotea. Está gráfica corresponde al ruido blanco.

```python
whiten = fftpack.ifft(whitespect)
whitenoise = whiten.real
plot(whitenoise)
```

A continuación se muestra una gráfica que contiene la señal original y la del ruido blanco

```python
plot(signal)
plot(whitenoise)
```

```python
formato = audio.Format('wav')
wn = audio.Sndfile('whitenoise.wav', 'w', formato, 1, sample_frequency)
wn.write_frames(whitenoise)
```

## Pink Noise

El ruido rosa es una señal en la que el espectro de potencias tiene una densidad potencial inversamente proporcional a la frecuencia de la señal. Se caracteriza porque tiene la misma distribución de potencia para cada octava. Para hallar el espectro del ruido rosa se divide sobre la frecuencia. Al graficar lo anterior, se puede observar cómo decrece en este factor. 

Tomado de: http://www.firstpr.com.au/dsp/pink-noise/#Filtering
http://www.scholarpedia.org/article/1/f_noise

```python
rosa=1/freq
plot(rosa)
```

```python
max_freq=(max(freq)).real
freqe=exp(freq/max_freq)
rosa1=whiten/freqe
plot(rosa1)
```

```python
pinkn = fftpack.ifft(rosa1)
pinknoise = pinkn.real
plot(signal,'g')
plot(pinkn, 'r')
```

```python
formato = audio.Format('wav')
wn = audio.Sndfile('pinknoise.wav', 'w', formato, 1, sample_frequency)
wn.write_frames(pinkn)
```
