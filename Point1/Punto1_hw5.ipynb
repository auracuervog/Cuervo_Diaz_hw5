```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
from mpl_toolkits.mplot3d import Axes3D
from scipy import fftpack
from scipy.misc.pilutil import Image
import scipy.misc
```

```python
def transform(imagefile):
    #Se abre la imagen y se convierte a escala de grises
    image = Image.open(imagefile)
    img = image.convert('L')
    
    #Se convierte la imagen en un arreglo
    img_a = scipy.misc.fromimage(img)
    
    #Se realiza la transformada rápida de fourier en 2D
    img_ft = fftpack.fft2(img_a, axes=(0, 1))

    return img_ft
```

Para realizar los filtros a las imágenes es necesario conocer la distribución de la transformada de Fourier. Así que ahora se usará una de las imágenes para verificar cómo están distribuidas las frecuencias de la transformada de Fourier

```python
img_ft = transform("colesterol-1.tif")

X=[]
Y=[]

for i in range(0,len(img_ft[:,0])):
    Y.append(i)
for i in range(0,len(img_ft[0,:])):
    X.append(i)
def ft(x,y):
    return img_ft[x,y]

plot(Y,ft(243,Y))
```

### Filtro Butterworth para altas frecuencias

A continuación se define la función que realiza el filtro de altas frecuencias usando el método de Butterworth para primer orden.

Como se observará más adelante, el filtro de altas frecuencias acentua los bordes de las imágenes.

```python
def highpass_filter(img_romp):
    #Se inicializan las columnas y filas de la imagen que resultará después del filtrado
    m = img_romp.shape[0]
    n = img_romp.shape[1]
    
    #Se crea la matriz de 0
    newimg = [[0 for x in range(m)] for x in range(n)]
    
    ycenter = m/2
    xcenter = n/2
    
    #Este es el radio al cual se realizará el filtro 
    filterr = 30
    t1 = 1 #Orden de la transformada de Butterworth
    t2 = 2*t1
    
    #Convolución
    for i in range(1,m):
        for j in range(1,n):
            #Distancia radial del punto que se está iterando al centro de la imagen
            r2 = ((i-ycenter)**2)+((j-xcenter)**2)
            r = np.sqrt(r2)
            if 0<r<filterr:
                newimg[i,j] = 1/(1+(filterr/r)**t2) 
    
    #La matriz resultante se convierte en imagen
    highfilter = scipy.misc.toimage(newimg)

    convolution = highfilter*img_romp
    magn_filter = abs(fftpack.ifft2(convolution))
    
    #Se genera la imagen
    high = scipy.misc.toimage(magn_filter)
    fig = figure(1, figsize=(6.5,4.5))
    imshow(high[:,:], cmap = cm.Greys_r)
    plt.title('Butterworth  highpass filter', fontsize=12)
    
```

### Filtro Butterworth para bajas frecuencias

A continuación se define la función que filtra las bajas frecuencias de la transformada de Fourier y genera la imagen resultante.

Como se observará más adelante, el filtro de bajas frecuencias reduce el ruido de la imagen y la suaviza.



```python
def lowpass_filter(img_romp):
    #Se inicializan las columnas y filas de la imagen que resultará después del filtrado
    m = img_romp.shape[0]
    n = img_romp.shape[1]
    
    #Se crea la matriz de 0
    newimg = [[0 for x in range(m)] for x in range(n)]
    
    ycenter = m/2
    xcenter = n/2
    
    #Este es el radio al cual se realizará el filtro 
    filterr = 30
    t = 2*filterr
    
    #Convolución:
    for i in range(1,m):
        for j in range(1,n):
            #Distancia radial del punto que se está iterando al centro de la imagen
            r2 = ((i-ycenter)**2)+((j-xcenter)**2)
            r = np.sqrt(r2)
            
            #Eliminando altas frecuencias
            if r>=filterr:
                newimg[i,j] = 1/((1+r/filterr)**t) 
                
    #La matriz resultante se convierte en imagen
    lowfilter = scipy.misc.toimage(newimg)

    convolution = lowfilter*img_romp
    magn_filter = abs(fftpack.ifft2(convolution))
    
    #Se genera la imagen
    new = scipy.misc.toimage(magn_filter)
    fig = figure(1, figsize=(6.5,4.5))
    imshow(new[:,:], cmap = cm.Greys_r)
    plt.title('Butterworth  lowpass filter', fontsize=12)
    
```
