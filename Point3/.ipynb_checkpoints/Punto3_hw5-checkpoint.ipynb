```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.misc
```

```python
xpos = []
ypos = []

positions = loadtxt("circulos.dat")

xpos = positions[:,0]
ypos = positions[:,1]

fig = plt.figure()
fig.set_size_inches(8,8)
scatter(xpos, ypos,s=1.5, alpha=0.5)
xlabel('Distancia en x')
ylabel('Distancia en y')        
```

```python
nx = 1024
ny = 1024

deltax = 6.0/nx
deltay = 6.0/ny

intensidad = [[0 for y in range(ny)]for x in range(nx)]

for i in range(nx):
   for j in range(ny):
    min_x = i*deltax
    max_x = min_x+deltax
    min_y = j*deltay
    max_y = min_y+deltay
    if (min_x<xpos[i]<max_x & min_y<ypos[i]<max_y):
        intensidad[i][j] +=1
```
