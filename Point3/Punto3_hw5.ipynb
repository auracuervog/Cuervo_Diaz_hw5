El objetivo de este ejercicio es que dadas las posiciones de n puntos en x-y que forman n círculos, se halle el radio promedio. Lo primero que realizaremos es graficar los puntos para conocer su distribución y comprobar que forman círculos.

```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.misc
from scipy.misc.pilutil import Image
from scipy import fftpack
```

```python
xpos = []
ypos = []

positions = loadtxt("circulos.dat")

xpos = positions[:,0]+3
ypos = positions[:,1]-3

fig = figure(1, figsize=(7.5,7.5))
scatter(xpos, ypos,s=1.5, alpha=0.5)
xlabel('Distancia en x')
ylabel('Distancia en y')
```

A continuación se genera una imagen a partir de la distribución de los puntos en el espacio.

```python
nx = 512
ny = 512

deltax = 6.0/nx
deltay = 6.0/ny

intensidad = [[0 for y in range(ny)]for x in range(nx)]

for i in range(len(xpos)):
    exis=int(trunc(abs(xpos[i])/deltax))
    ye=int(trunc(abs(ypos[i])/deltay))
    intensidad[ye][exis]+=1
    
    #Se genera la imagen
fig = figure(1, figsize=(7.5,5.5))
im = scipy.misc.toimage(intensidad)
plt.imshow(im)
im.save('imagen.png')
```
