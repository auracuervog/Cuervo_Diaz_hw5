```python
nx = 512
ny = 512

deltax = 6.0/nx
deltay = 6.0/ny

intensidad = [[0 for y in range(ny)]for x in range(nx)]

for i in range(len(xpos)):
    exis=int(trunc(abs(xpos[i])/deltax))
    ye=int(trunc(abs(ypos[i])/deltay))
    intensidad[ye][exis]+=1
    
    #Se genera la imagen
im = scipy.misc.toimage(intensidad)
im.show()
im.save('imagen.png')
```
